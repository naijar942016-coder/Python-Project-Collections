```python
# 1.String Concatenation
string1 = "Hello Zara "
name = input("Enter your Name:",)

# Step 1: concatenate
result = string1 + name
print(result)

# Step 2: add welcome message
string3 = ", welcome to Python programming"
result = result + string3
print(result)

#2.Slicing and indexing
s = "Hello Zara, welcome to Python programming"

print(s[0])         # a. first character
print(s[-1])        # b. last character
print(s[:5])        # c. first 5 characters
print(s[-11:])      # d. last 11 characters
print(s[::-1])      # e. reverse
print(s[23:29])     # f. "Python"

#3.Sting Methods
strM = "Python beginner tutorial"

# a. uppercase
print(strM.upper())


#4.Tuples
t1 = (10, 20, 30)
t2 = (40, 50, 60)

# a. concatenate
t_combine = t1 + t2
print("Combined:", t_combine)

# b. repeat 3 times
print("Repeated:", t_combine * 3)

# c. 3rd element (index 2)
print("3rd element:", t_combine[2])

# d. first three elements
print("First 3:", t_combine[:3])

# e. last three elements
print("Last 3:", t_combine[-3:])
```

```python


# 1a. Create age_list with five integers
age_list = [24, 25, 26, 27, 28]
print("Original age_list:", age_list)

# 1b. Create name_list with five strings
name_list = ["Naija", "Priya", "Divya", "Meena", "Kavya"]
print("Original name_list:", name_list)

#--------------------------------------------
# 2. List Operations / Modifications
#--------------------------------------------

# 2a. Append "Yazhini" to name_list
name_list.append("Yazhini")
print("\nAfter append:", name_list)

# 2b. Insert 30 at index 2 in age_list
age_list.insert(2, 30)
print("After insert at index 2:", age_list)

# 2c. Remove "Yazhini" from name_list
name_list.remove("Yazhini")
print("After remove Yazhini:", name_list)

# 2d. Pop the last element from age_list
popped = age_list.pop()
print("Popped element:", popped)
print("After pop:", age_list)

# 2e. Extend age_list with [29, 30, 26]
age_list.extend([29, 30, 26])
print("After extend:", age_list)

# 2f. Sort age_list in descending order
age_list.sort(reverse=True)
print("After sort (descending):", age_list)

# 2g. Max, Min, and Sum of age_list
print("Max age:", max(age_list))
print("Min age:", min(age_list))
print("Sum of ages:", sum(age_list))

#--------------------------------------------
# 3. Accessing List Elements
#--------------------------------------------

# 3a. First element of name_list
print("\nFirst element:", name_list[0])

# 3b. Last element of name_list
print("Last element:", name_list[-1])

# 3c. Elements from index 2 to 4
print("Elements index 2 to 4:", name_list[2:5])

# 3d. name_list in reverse order
print("Reversed name_list:", name_list[::-1])


#--------------------------------------------
# SECTION 2: DICTIONARY
#--------------------------------------------

# a. Create student_marks dictionary
student_marks = {
    "Naija"  : 88,
    "Priya"  : 75,
    "Divya"  : 91,
    "Meena"  : 63,
    "Kavya"  : 79
}
print("\n--- Dictionary ---")
print("student_marks:", student_marks)

# b. Access and print a specific student's mark
print("Mark of Divya:", student_marks["Divya"])

# c. Add new student "Janani" with mark 80
student_marks["Janani"] = 80
print("After adding Janani:", student_marks)

# d. Update an older student's mark to 82
student_marks["Priya"] = 82
print("After updating Priya's mark:", student_marks)

# e. keys(), values(), items()
print("\nKeys   :", list(student_marks.keys()))
print("Values :", list(student_marks.values()))
print("Items  :")
for key, value in student_marks.items():
    print("  ", key, "->", value)


#--------------------------------------------
# SECTION 3: SETS
#--------------------------------------------

print("\n--- Sets ---")

# a. Create my_set from list with duplicates
my_set = set(['a','e','i','o','u','a','a','i'])
print("my_set:", my_set)



# b. Attempt to change my_set[4] = 's'
print("Attempting my_set[4] = 's'...")
try:
    my_set[4] = 's'
except TypeError as e:
    print("Error:", e)
    

# c. Create set1 and set2
set1 = {1, 3, 5, 7, 9}
set2 = {2, 3, 5, 8, 10}
print("set1:", set1)
print("set2:", set2)

# d. Union and Intersection
print("Union       :", set1 | set2)
print("Intersection:", set1 & set2)


#--------------------------------------------
# SECTION 4: OPERATORS & CONDITIONAL STATEMENTS
#--------------------------------------------

print("\n--- Performance Category ---")

score = int(input("Enter your score (0 to 10): "))

if score < 0 or score > 10:
    print("Invalid score! Please enter a value between 0 and 10.")
elif score > 7:
    print("Performance Category: Above Average")
    print("Great work! Keep up the excellent performance and aim even higher!")
elif score >= 4:
    print("Performance Category: Average")
    print("Good effort! Push yourself a little more to reach the next level.")
else:
    print("Performance Category: Below Average")
    print("Need to improve your performance. Consistent practice will lead to better results!")
```

```python

#Using Function


def calculate_bmi(weight, height):
    """Returns BMI = weight(kg) / height(m)^2"""
    return weight / (height ** 2)

weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))

bmi = calculate_bmi(weight, height)
print(f"Your BMI is: {bmi:.2f}")
```
